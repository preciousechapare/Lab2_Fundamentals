```python
# APPENDIX SETUP (put this at the top)

ENGINEER = "ECHAPARE"
STUDENT_ID = "TUPM-25-3619"

SEED_NUM = int(STUDENT_ID[-1])

first_letter = ENGINEER[0]
byte_val = ord(first_letter)

print("ENGINEER:", ENGINEER)
print("STUDENT_ID:", STUDENT_ID)
print("SEED_NUM:", SEED_NUM)
print("ASCII (byte_val):", byte_val)
```

```python
target_name = "ECHAPARE".upper()   # <-- REPLACE YOUR_SURNAME
SEED_NUM = int(STUDENT_ID[-1])

vowels = "AEIOU"
redacted_name = ""

for ch in target_name:
    if ch in vowels:
        redacted_name += str(SEED_NUM)
    else:
        redacted_name += ch

print("Original Surname:", target_name)
print("SEED_NUM Used:", SEED_NUM)
print("Redacted Output:", redacted_name)
```

```python
# EXERCISE 2: VERIFICATION HASH
# Formula: verification_hash = (ASCII * SEED_NUM) - loops

A = byte_val   # 69
B = SEED_NUM   # 9
C = 12         # loops from your Cell 11 (SEED_NUM + 3 = 12)

verification_hash = (A * B) - C

print("First Letter ASCII (A):", A)
print("SEED_NUM (B):", B)
print("Total Loops (C):", C)
print("Verification Hash:", verification_hash)
```

```python
# EXERCISE 3: PASSWORD LOOP

correct_password = "********"   # keep exactly like this unless instructor gave a real one
attempts = 0

while True:
    guess = input("Type password: ")
    attempts += 1
    if guess == correct_password:
        break
    print("Wrong password.")

print("Correct Password Set:", correct_password)
print("Attempts Before Success:", attempts - 1)
print("Attempts Total:", attempts)
```
